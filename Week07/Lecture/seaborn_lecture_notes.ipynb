```python
titanic = pd.read_csv("http://bit.ly/kaggletrain")
titanic.head()
```

```python
titanic.sample(5, replace=True)
```

```python
titanic["Name"].sample(5, replace=True)
```

```python
for i in range(1000):
    res = titanic.sample(5, replace=True)
    indices = res.index
    if len(indices.unique()) != 5:
        print(indices)
```

```python
for i in range(1000):
    res = titanic.sample(5, replace=False)
    indices = res.index
    if len(indices.unique()) != 5:
        print(indices)
```

```python
res = []
for i in range(1000):
    temp_res = titanic["Fare"].sample(100, replace=True).mean()
    res.append(temp_res)

res.sort()
perc = 95
idx = int(((100 - perc) / 2) / 100 * len(res))
thr1, thr2 = res[idx], res[-idx]
sns.displot(res)
plt.plot([thr1, thr1], [0, 120], color="red", linewidth=3, linestyle="dashed", label=format(thr1, ".2f"))
plt.plot([thr2, thr2], [0, 120], color="purple", linewidth=3, linestyle="dashed", label=format(thr2, ".2f"))
plt.legend()
plt.show()
```
