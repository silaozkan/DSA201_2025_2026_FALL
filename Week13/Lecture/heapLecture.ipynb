```python
def insert(heap, val): # assuming heap is a max-heap
    heap.append(val)
    cur_idx = len(heap) - 1
    parent_idx = (cur_idx - 1) // 2
    while parent_idx >= 0 and heap[parent_idx] < heap[cur_idx]:
        temp = heap[parent_idx]
        heap[parent_idx] = heap[cur_idx]
        heap[cur_idx] = temp

        cur_idx = parent_idx
        parent_idx = (cur_idx - 1) // 2
```

```python
myHeap = [22, 11, 19, 3, 5, 1]
insert(myHeap, 88)
print(myHeap)
```
